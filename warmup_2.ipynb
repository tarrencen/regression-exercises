```python
energy_lvl5 = len(df[df['Energy Level'] == 5])
print(f'There are {energy_lvl5} breeds that have an energy level of 5.')
```

```python
unique_coats = df['Coat Type'].nunique()
print(f'There are {unique_coats} different coat types.')
```

```python
long_coats = df[df['Coat Length'] == 'Long']
avg_lc_drool = long_coats['Drooling Level'].mean()
print(f'Average drooling level for doggos with long coats is: {avg_lc_drool:.3f}')
```
